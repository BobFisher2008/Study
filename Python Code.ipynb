```python
# Calculate minutely order flows (net MO within each minute)
order_flow = np.diff(MO[:, 0])  # Assuming MO has shape (num_samples, 1)

# Calculate the autocorrelation function (ACF) up to 20 lags
lags = 20
acf_order_flow = np.correlate(order_flow, order_flow, mode='full')
acf_order_flow = acf_order_flow / np.max(acf_order_flow)  # Normalize ACF values

# Plot the ACF of order flows
plt.acorr(order_flow, maxlags=lags)
plt.title('Autocorrelation Function of Minutely Order Flows')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```

```python
import statsmodels.api as sm

# Calculate minutely returns based on mid-price
midprice_returns = midprice[1:] / midprice[:-1] - 1
midprice_returns = np.insert(midprice_returns, 0, np.nan)

# Calculate minutely order flows (net MO within each minute)
order_flow = np.diff(MO[:, 0])  # Assuming MO has shape (num_samples, 1)

# Create a lagged return variable
lagged_return = np.roll(midprice_returns, 1)
lagged_return[0] = np.nan  # Set the first value to NaN since there is no lagged return for the first observation

# Make sure all arrays have the same length
min_length = min(len(lagged_return), len(order_flow), len(midprice_returns))
lagged_return = lagged_return[:min_length]
order_flow = order_flow[:min_length]
midprice_returns = midprice_returns[:min_length]

# Create a DataFrame with the variables
regression_data = pd.DataFrame({
    'Lagged_Return': lagged_return,
    'Order_Flow': order_flow,
    'Minutely_Return': midprice_returns
})

# Drop rows with missing values
regression_data = regression_data.dropna()

# Perform linear regression
X = sm.add_constant(regression_data[['Lagged_Return', 'Order_Flow']])
y = regression_data['Minutely_Return']
model = sm.OLS(y, X).fit()

# Print regression results
print(model.summary())
```

```python
window_limit = 1800
t = t[:window_limit]
order_flows = order_flows[:window_limit]
prices = prices[:window_limit]

# Calculate cumulative order flows for the first 1800 seconds
cumulative_order_flows = np.cumsum(order_flows)

# Initialize positions and returns
positions = np.zeros(len(t))
returns = np.zeros(len(t))

# Define the strategy: long if cumulative order flows > 0, short if < 0
for i in range(1, len(t)):
    if cumulative_order_flows[i-1] > 0:
        positions[i] = 1  # Long
    elif cumulative_order_flows[i-1] < 0:
        positions[i] = -1  # Short
    else:
        positions[i] = positions[i-1]  # Hold previous position

    # Calculate returns based on the midprice change
    returns[i] = (midprice[i] - midprice[i-1])/midprice[i]

# Calculate cumulative returns
cumulative_returns = np.cumsum(returns)

# Plot cumulative order flows and cumulative returns within the first 1800 seconds
plt.figure(figsize=(15, 8))

plt.subplot(3, 1, 1)
plt.plot(t, cumulative_order_flows)
plt.title('Cumulative Order Flows (Previous 30 Minutes)')
plt.xlabel('Time (seconds from trading start)')
plt.ylabel('Cumulative Order Flows')
plt.xlim(0, 1800)

plt.subplot(3, 1, 3)
plt.plot(t, cumulative_returns)
plt.title('Cumulative Returns of Trading Strategy')
plt.xlabel('Time (seconds from trading start)')
plt.ylabel('Cumulative Returns')
plt.xlim(0, 1800)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
p = 0.2  # initial probability that V = V_H
alpha = 0.8  # proportion of informed traders
V_H = 2
V_L = 1
trades = 50  # number of trades to simulate
spread = 0.05  # a small spread to distinguish ask and bid prices

# Probabilities for the likelihood function
P_buy_given_VH = 0.5 + 0.5 * alpha
P_buy_given_VL = 0.5 - 0.5 * alpha
P_sell_given_VH = 0.5 - 0.5 * alpha
P_sell_given_VL = 0.5 + 0.5 * alpha

# Function to update beliefs using Bayes' theorem
def update_belief(P_t, trade, alpha):
    if trade == 'buy':
        P_t1 = (P_t * P_buy_given_VH) / (P_t * P_buy_given_VH + (1 - P_t) * P_buy_given_VL)
    else:  # trade == 'sell'
        P_t1 = (P_t * P_sell_given_VH) / (P_t * P_sell_given_VH + (1 - P_t) * P_sell_given_VL)
    return P_t1

# Initialize belief
P_t = p
P_values = [P_t]

# Simulate trades
np.random.seed(42)
trades_sequence = []
for _ in range(trades):
    if np.random.rand() < 0.5:
        trades_sequence.append('buy')
    else:
        trades_sequence.append('sell')

# Update beliefs and calculate ask/bid prices
A_values = []
B_values = []

for trade in trades_sequence:
    P_t = update_belief(P_t, trade, alpha)
    P_values.append(P_t)
    A_t = P_t * V_H + (1 - P_t) * V_L + spread  # Ask price with a spread
    B_t = P_t * V_H + (1 - P_t) * V_L - spread  # Bid price with a spread
    A_values.append(A_t)
    B_values.append(B_t)

# Plot A and B against time
time = np.arange(trades)
plt.figure(figsize=(12, 6))
plt.plot(time, A_values, label='Ask Price (A)', color='blue')
plt.plot(time, B_values, label='Bid Price (B)', color='red')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Ask and Bid Prices Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
